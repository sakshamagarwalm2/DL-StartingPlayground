#Creating a functional model using keras

importing libreries

```python
import pandas as pd
import numpy as np

np.random.seed(1212)

import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
```

```python
df_train = pd.read_csv('/content/sample_data/mnist_train_small.csv')
df_test = pd.read_csv('/content/sample_data/mnist_test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_features = df_train.iloc[:,1:785]
df_label = df_train.iloc[:,0]

X_test = df_test.iloc[:,0:784]

print(X_test.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_cv, y_train, y_cv = train_test_split(df_features, df_label, test_size=0.2, random_state=1212)

print(X_train.shape,X_cv.shape)
```

```python
X_train = X_train.reshape(15999,784)
X_cv = X_cv.reshape(4000,784)

# X_test = X_test.values.reshape(9999,784)
```

```python
#feature Normalization
X_train = X_train.astype('float32')/255
X_cv = X_cv.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
#convert labels to one hot encoder
y_train = keras.utils.to_categorical(y_train,10)
y_cv = keras.utils.to_categorical(y_cv,10)
```

```python
#Input parameters
n_input = 784
n_h_1 = 300
n_h_2 = 100
n_h_3 = 100
n_h_4 = 200
n_d =10
```

```python
# creating layers
inp = Input(shape=(784,))
x = Dense(n_h_1, activation='relu', name="n_h_1")(inp)
x = Dense(n_h_2, activation='relu', name="n_h_2")(x)
x = Dense(n_h_3, activation='relu', name="n_h_3")(x)
x = Dense(n_h_4, activation='relu', name="n_h_4")(x)
output = Dense(n_d, activation='softmax', name="out")(x)
```

```python
# creating model
model = Model(inp, output)
model.summary()
```

```python
#Inserting hyperparameters
learning_rate = 0.1
training_epochs = 20
batch_size = 100
sgd = optimizers.SGD(lr=learning_rate)
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
print(X_train.shape)
```

```python
h1 = model.fit(X_train, y_train,
               batch_size = 100,
               epochs= 20,
               verbose=2,
               validation_data=(X_cv,y_cv))
```
